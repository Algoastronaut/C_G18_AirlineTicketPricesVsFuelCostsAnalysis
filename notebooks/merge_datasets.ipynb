# Data Merging 

Merging the 6 datasets into three to on logical pairs (I think!!!!!)

```python
import pandas as pd
import os

print('Current directory:', os.getcwd())
```

### 1. Macro Economic & Events Merge
Merging `oil_jet_fuel_prices.csv` with `conflict_oil_events.csv`.

```python
df_oil = pd.read_csv('../raw/oil_jet_fuel_prices.csv')
df_events = pd.read_csv('../raw/conflict_oil_events.csv')

df_events['month'] = df_events['event_date'].str[:7]
df_macro = pd.merge(df_oil, df_events, on=['month', 'conflict_phase'], how='left', suffixes=('', '_event'))
df_macro.to_csv('1_macro_oil_and_events.csv', index=False)
print(f"Pair 1 shape: {df_macro.shape}")
```

### 2. Financial & Route Cost Merge
Merging `route_cost_impact.csv` with `airline_financial_impact.csv`.

```python
df_route = pd.read_csv('../raw/route_cost_impact.csv')
df_fin = pd.read_csv('../raw/airline_financial_impact.csv')

keys_route_fin = ['month', 'conflict_phase', 'airline']
df_route_fin = pd.merge(df_route, df_fin, on=keys_route_fin, how='left', suffixes=('', '_financial'))
df_route_fin = df_route_fin.loc[:, ~df_route_fin.columns.duplicated()]
df_route_fin.to_csv('2_route_impacts_and_financials.csv', index=False)
print(f"Pair 2 shape: {df_route_fin.shape}")
```

### 3. Passenger Pricing & Fuel Surcharges Merge
Merging `airline_ticket_prices.csv` with `fuel_surcharges.csv`.

```python
df_tickets = pd.read_csv('../raw/airline_ticket_prices.csv')
df_surcharges = pd.read_csv('../raw/fuel_surcharges.csv')

def map_km(km):
    if km <= 1500:
        return 'Up to 1500 km'
    elif km <= 4500:
        return '1501–4500 km'
    elif km <= 9000:
        return '4501–9000 km'
    else:
        return 'Over 9000 km'
        
df_tickets['km_range'] = df_tickets['avg_route_km'].apply(map_km)

keys_tickets_surcharges = ['month', 'conflict_phase', 'airline', 'iata_code', 'country', 'region', 'km_range']
df_tickets_surcharges = pd.merge(df_tickets, df_surcharges, on=keys_tickets_surcharges, how='left', suffixes=('', '_surcharge_policy'))
df_tickets_surcharges = df_tickets_surcharges.loc[:, ~df_tickets_surcharges.columns.duplicated()]
df_tickets_surcharges.to_csv('3_ticket_prices_and_surcharges.csv', index=False)
print(f"Pair 3 shape: {df_tickets_surcharges.shape}")
```
